# Satellite Image Segmentation of Structures with tensorflow

# 使用tensorflow分割卫星影像上的建筑物

该解决方案主要参照：https://github.com/tensorflow/models/blob/master/samples/outreach/blogs/segmentation_blogpost/image_segmentation.ipynb ，除了数据准备方式不同，其他的数据增强和模型编写等步骤基本和原文一致。

数据来源为一副大约24000x23000的遥感影像image，手动勾绘出label图片，使用cv2将image和label分别分割为256x256大小的jpg格式（原始影像为tiff格式，tensorflow原生不支持，所以转为jpg格式）的小图片，实际分割成约8000个样本。
基于这8000个样本做了一次训练，效果不好，估计原因是：
1. 手动勾绘时漏绘，即影像image上有建筑物的地方没有在label中有体现
2. 有建筑物的图片只占样本数量的一小部分（1500/8000）
基本这次结果，只将在label中有建筑物的image和label图片筛选出来，使用这些图片进行训练，最终结果还算可以。
